### Comment Writing Engine

#### an application to help teachers write better comments in less time



```python
#Abby
import csv
import random
choice_file = 'Choice.csv'

bank = {}
f = open(choice_file,'r')
reader = csv.reader(f)
header = next(reader)

for row in reader:
    tag = row[0].strip().lower()
    sentence = row[1].strip()
    if len(sentence) >= 2 and sentence[0] == '"' and sentence[-1] == '"':
        sentence = sentence[1:-1]
    if tag not in bank:
        bank[tag] = []
    bank[tag].append(sentence)
f.close()

data_file = 'Data .csv'
f = open(data_file,'r')
reader = csv.reader(f)
header = next(reader)

#Abby and Ruby
def make_comment(row, bank):
    """
    Takes one row of student data and the sentence bank, 
    then returns (name, comment) for that student.
    """
    name = row[0].strip()
    sem = row[1].strip()
    proj = row[2].strip()
    s1 = row[3].strip().lower()
    s2 = row[4].strip().lower()
    g1 = row[5].strip().lower()
    g2 = row[6].strip().lower()
    evidence = row[7].strip()
    note = row[8].strip()
    
    parts = []
    parts.append(name+", your semester grade is "+sem+", and your project grade is "+proj+".")
    if s1 in bank:
        parts.append(random.choice(bank[s1]))
    if s2 in bank:
        sent2 = random.choice(bank[s2])
        if sent2 not in parts:
            parts.append(sent2)
    if evidence != "":
        if evidence[-1] == ".":
            evidence = evidence[:-1]
        parts.append("One specific highlight of the semester was "+evidence+".")
    if g1 in bank:
        parts.append(random.choice(bank[g1]))
    if g2 in bank:
        sent4 = random.choice(bank[g2])
        if sent4 not in parts:
            parts.append(sent4)
    if note != "":
        parts.append(note)
    parts.append("Keep it up next semester.")
   
    comment = " ".join(parts)
    return name,comment

#Ruby    
def write_comment(name, comment):
    """
    Saves the comment into a txt file (example: Ruby.txt)
    """
    out = open(name+'.txt','w')
    out.write(comment)
    out.close()
    
for row in reader:
    name,comment=make_comment(row,bank)
    write_comment(name,comment)
```
